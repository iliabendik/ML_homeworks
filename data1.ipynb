```python
from sklearn import tree
```

```python
test_iris = pd.read_csv('test_iris.csv')
X_test, y_test = test_iris[test_iris.columns[:-1]].values, test_iris[test_iris.columns[-1]].values
train_iris = pd.read_csv('train_iris.csv')
X_train, y_train = train_iris[train_iris.columns[:-1]].values, train_iris[train_iris.columns[-1]].values
score_dataset = pd.DataFrame()
for i in range(1, 100):
    np.random.seed(0)
    my_awesome_tree = tree.DecisionTreeClassifier(criterion='entropy', max_depth = i)
    my_awesome_tree.fit(X_train, y_train)
    score_dataset = score_dataset.append(pd.DataFrame({'max_depth': [i], 'training': [my_awesome_tree.score(X_train, y_train)], 'testing': [my_awesome_tree.score(X_test, y_test)]}))
score_dataset
```
